```python
import matplotlib.pyplot as plt
import chart_studio.plotly as py
import numpy as np
%matplotlib inline
```

# Functional method

```python
x = np.linspace(0,5,10)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,1,1])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Bigger Axis')

axes2.plot(y,x)
axes2.set_title('Smaller Axis')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title("Left")

axes[1].plot(y,x)
axes[1].set_title("Right")

plt.tight_layout()
```

# Figure size and DPI

```python
fig = plt.figure(figsize=(10,2)) # figure Size

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,2)) # figure Size

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_fig.jpg')
```

```python
fig = plt.figure(figsize=(5,4)) # figure Size

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3,label='Y cubed')

ax.legend()
```

# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=5,alpha=0.5,ls='--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green',marker='o')
```
