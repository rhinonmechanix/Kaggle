## Categorical Plots

```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.mean)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

## Matrix Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True)
```

```python
flights
```

```python
ft = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(ft,cmap='crest')
"""
cmap = 'flare','rocket','mako','crest'
"""
```

```python
sns.heatmap(ft,cmap='mako',linecolor='white',linewidth=2)
```

```python
sns.clustermap(ft)
```

## Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris["species"].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
a = sns.FacetGrid(col='day',row='smoker',data=tips)
a.map(sns.distplot,'total_bill')
```
